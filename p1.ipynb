```python
def search(data):
    X = input("Enter search element: ")

    for i in range(len(data)):
        if data[i] == X:
            print(f"Element found at index {i}")
            return

    print("Element not found")


def sort(data):
    for i in range(len(data)):
        flag = False
        for j in range(len(data) - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
                flag = True
        if not flag:
            return


def update(data, key):
    if key not in data:
        print("Key not found")
    else:
        X = input("Enter a new value: ")
        data[key] = X


List_ = []


def list_op():
    print("List Operations:")
    print("1. Insert\n2. Delete\n3. Update\n4. Search\n5. Sorting\n6. Display\n7. Return to main")

    while True:
        ch = int(input("Enter Choice: "))

        if ch == 1:
            X = int(input("Enter element: "))
            List_.append(X)
        elif ch == 2:
            ele = int(input("Enter element to be deleted: "))
            try:
                List_.remove(ele)
            except:
                print(f"Element {ele} not found")
        elif ch == 3:
            index = int(input("Enter index: "))
            update(List_, index)
        elif ch == 4:
            search(List_)
        elif ch == 5:
            sort(List_)
        elif ch == 6:
            print(List_)
        elif ch == 7:
            break
        else:
            print("Invalid Choice")


Tuple_ = ()


def tuple_op():
    print("Tuple Operations:")
    print("1. Insert\n2. Delete\n3. Update\n4. Search\n5. Display\n6. Return to main")

    while True:
        ch = int(input("Enter the choice: "))

        if ch == 1:
            ele = int(input("Enter element: "))
            Tuple_ += (ele,)
        elif ch == 2:
            print("Cannot delete from tuple.")
        elif ch == 3:
            print("Cannot update in tuple.")
        elif ch == 4:
            search(Tuple_)
        elif ch == 5:
            print(Tuple_)
        elif ch == 6:
            break
        else:
            print("Invalid choice")


myset = set()


def set_op():
    print("Set Operations:")
    print("1. Add\n2. Remove\n3. Search\n4. Display\n5. Return to main")

    while True:
        ch = int(input("Enter the choice: "))

        if ch == 1:
            myset.add(input("Enter Element: "))
        elif ch == 2:
            ele = input("Enter element to remove: ")
            if ele in myset:
                myset.remove(ele)
            else:
                print("Element not found")
        elif ch == 3:
            search(myset)
        elif ch == 4:
            print(myset)
        elif ch == 5:
            break
        else:
            print("Invalid Choice")


mydict = {}


def dict_op():
    print("Dictionary Operations:")
    print("1. Insert\n2. Update\n3. Delete\n4. Search\n5. Display\n6. Return to main")

    while True:
        ch = int(input("Enter the Choice: "))

        if ch == 1:
            key = input("Enter Key: ")
            value = input("Enter Value: ")
            mydict[key] = value
        elif ch == 2:
            key = input("Enter Key to Update: ")
            update(mydict, key)
        elif ch == 3:
            key = input("Enter Key to Delete: ")
            if key in mydict:
                del mydict[key]
            else:
                print("Key not found")
        elif ch == 4:
            search(mydict.values())
        elif ch == 5:
            print(mydict)
        elif ch == 6:
            break
        else:
            print("Invalid Choice")


while True:
    print("1. List\n2. Tuple\n3. Set\n4. Dictionary\n5. Exit")

    choice = int(input("Enter the Choice: "))

    if choice == 1:
        list_op()
    elif choice == 2:
        tuple_op()
    elif choice == 3:
        set_op()
    elif choice == 4:
        dict_op()
    elif choice == 5:
        print("Exiting program...")
        break
    else:
        print("Invalid choice")
```
