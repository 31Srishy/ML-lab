```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate some sample data
np.random.seed(42)
cluster_1 = np.random.randn(100, 2) + np.array([0, -3])
cluster_2 = np.random.randn(100, 2) + np.array([3, 3])
cluster_3 = np.random.randn(100, 2) + np.array([-3, 3])
X = np.vstack((cluster_1, cluster_2, cluster_3))

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Get the cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_
iterations = kmeans.n_iter_

# Output the calculations
print("K-means Clustering Results:")
print("Cluster Centers:\n", centers)
print("Number of Iterations:", iterations)
print("Labels:\n", labels)

# Plot the clustered data
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate simpler sample data
np.random.seed(42)
simple_data = np.array([
    [1, 2], [2, 1], [3, 2], [4, 1], [5, 2], # Cluster 1
    [8, 8], [8, 9], [9, 8], [9, 9],         # Cluster 2
    [1, 8], [2, 9], [1, 9], [2, 8]          # Cluster 3
])

# Function to plot the data
def plot_clusters(X, labels, title=''):
    plt.scatter(X[:, 0], X[:, 1], c=labels, s=100, cmap='viridis', edgecolors='black')
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

# Single-linkage agglomerative clustering
single_linkage = AgglomerativeClustering(n_clusters=3, linkage='single')
single_labels = single_linkage.fit_predict(simple_data)
print("Single-Linkage Agglomerative Clustering Labels:\n", single_labels)
plot_clusters(simple_data, single_labels, title='Single-Linkage Agglomerative Clustering')

# Complete-linkage agglomerative clustering
complete_linkage = AgglomerativeClustering(n_clusters=3, linkage='complete')
complete_labels = complete_linkage.fit_predict(simple_data)
print("Complete-Linkage Agglomerative Clustering Labels:\n", complete_labels)
plot_clusters(simple_data, complete_labels, title='Complete-Linkage Agglomerative Clustering')

# Plot the dendrogram for single-linkage
Z_single = linkage(simple_data, 'single')
print("Single-Linkage Dendrogram Linkage Matrix:\n", Z_single)
plt.figure()
dendrogram(Z_single)
plt.title('Dendrogram (Single-Linkage)')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()

# Plot the dendrogram for complete-linkage
Z_complete = linkage(simple_data, 'complete')
print("Complete-Linkage Dendrogram Linkage Matrix:\n", Z_complete)
plt.figure()
dendrogram(Z_complete)
plt.title('Dendrogram (Complete-Linkage)')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```
